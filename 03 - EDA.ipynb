# 03 - EDA Notebook
The intent of this notebook is to explore the relationships that the Clustering Models defined by viewing our data in a pivot table format, and plotting the results to visualize our findings.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### KMeans Model with Five Clusters
##### Cluster 0

```python
# Reading in Cluster 0 CSV
cluster_0_df = pd.read_csv('./state_employment/KMeans-5_Cluster-0_employment.csv', index_col='date')
cluster_0_df.head()
```

##### Setting up and plotting from a pivot table
To update which Metric to compare to, simply change the `values='unemployment rate'` to a different metric.

```python
pivot = cluster_0_df.pivot_table(values='unemployment rate', index=cluster_0_df.index, columns='State')

pivot.plot(legend=False);
```

##### Returns the Mean for metric set on the `pivot` variable

```python
pivot.T.mean().plot();
```

```python
pivot.columns
```
