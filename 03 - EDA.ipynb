```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./State Employment/KMeans-5_Cluster-0_employment.csv', index_col='date')

pivot = df.pivot_table(values='unemployment rate', index=df.index, columns='State')

pivot.plot(legend=False)
```

```python
pivot.T.mean().plot()
```

```python
pivot.columns
```
